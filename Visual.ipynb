```python
#import necessary libraries for the project

import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import warnings
warnings.filterwarnings('ignore')

import hvplot.pandas

#!pip install --upgrade param (USED TO INSTALL PANEL INITALLY)

```

```python
#import dataset

df = pd.read_csv("co2-data.csv")

df
```

```python
#Checking rows with country as 'North America'

df[df['country'] == 'North America']
```

```python
#checking rows with column with country as 'world'

df[df['country'] == 'World']
```

```python
#filling the 'NAN' values in dataset with '0'

df = df.fillna(0)

df['gdp_per_captia'] = np.where(df['population'] != 0, df['gdp']/df['population'],0)

df.head(2)
```

```python
# make dataframe pipline interactive

idf = df.interactive()
```

```python
# Creating a Panel widget of slider for years

year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2020, step=5, value=1800)
year_slider
```

```python
# Radio buttons for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['co2', 'co2_per_capita',],
    button_type='success'
)
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x = 'year', by='country', y=yaxis_co2,line_width=2, title="CO2 emission by continent")
co2_plot
```



> - <font color='blue'> ***Creating `Table` as one of visualization for dashboard*** </font>



```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
co2_table
```





> - <font color='blue'> ***Creating `Scatter plot` as one of visualization for dashboard*** </font>





```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[(idf.year == year_slider) & (~ (idf.country.isin(continents)))]
    .groupby(['country', 'year', 'gdp_per_captia'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_captia', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500)
co2_vs_gdp_scatterplot
```


> - <font color='blue'> ***Creating `Bar plot` as one of visualization for dashboard*** </font>





```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='CO2 source by continent')
co2_source_bar_plot
```


> - <font color='blue'> ***Creating final `dashboard`*** </font>





```python
#Layout using Template
# 5th line of this code you can use your image and paste the path of same.

template = pn.template.FastListTemplate(
    title='Dashboard using Python', 
    sidebar=[pn.pane.Markdown("# CO2 Emissions and Climate Change"), 
             pn.pane.Markdown("#### Carbon dioxide emissions are the primary driver of global climate change. Its widely recognised that to avoid the worst impacts of climate change, the world needs to urgently reduce emissions. But, how this responsibility is shared between regions, countries, and individuals has been an endless point of contention in international discussions."), 
             pn.pane.PNG('C:/Users/Siny/Pictures/20210608_111904_0000_2.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             year_slider],
    main=[pn.Row(pn.Column(yaxis_co2, 
                           co2_plot.panel(width=700), margin=(0,25)), 
                 co2_table.panel(width=700)), 
          pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
    accent_base_color="#ADD8E6",
    header_background="#88d8b0",
)
# template.show()
template.servable();
```


> - <font color='blue'> ***`Running the dashboard`*** </font>



```python
pn.serve(template)

#running this cell with pop up a new window/tab of the dashboard
```
